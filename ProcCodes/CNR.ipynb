```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
nameC="In-line"
nameV="TPMMAS"
nameD="Aire"
nameE="R6"
nameDa="PData"
nameGE="RAWSN"
nameGI="ThS"
nameRAW=nameC+"/"+nameV+"/"+nameD+"/"+nameE+"/"+nameGE+".txt"
nameTh=nameC+"/"+nameV+"/"+nameD+"/"+nameE+"/"+nameDa+"/"+nameGI+".txt"
RAW=np.genfromtxt(nameRAW)
Th=np.genfromtxt(nameTh)
Npixeles=256
```

```python
HistRAW=plt.hist(np.ravel(RAW),bins=50,color="blue")
```

```python
def PI(name,inicial,final,h,y):
    P=np.zeros(y-h)
    for i in range(inicial,final):
        P=P+name[i,h:y]
    mean=P/(final-inicial)
    return mean
```

```python
#X1=85      #R1
#X2=105
#X3=112
#X4=143
#X5=150
#X6=170
#X1=80       #R2
#X2=99
#X3=106
#X4=149
#X5=157
#X6=176
#X1=71      #R3
#X2=90
#X3=98
#X4=157
#X5=164
#X6=185
#X1=64      #R4
#X2=83
#X3=90
#X4=165
#X5=172
#X6=192
#X1=53      #R5
#X2=73
#X3=81
#X4=174
#X5=182
#X6=202
X1=28      #R6
X2=46
X3=56
X4=200
X5=208
X6=228
Y1=0
Y2=255
fig, ax = plt.subplots(figsize=(7, 7))
u=ax.imshow(Th, cmap="bone",vmin=-0.5,vmax=1.5)
ax.vlines(x=X1, ymin=Y1, ymax=Y2,linestyle="--", color="red",linewidth=1.5)
ax.vlines(x=X2, ymin=Y1, ymax=Y2,linestyle="--", color="red",linewidth=1.5)
ax.vlines(x=X3, ymin=Y1, ymax=Y2,linestyle="--", color="green",linewidth=1.5)
ax.vlines(x=X4, ymin=Y1, ymax=Y2,linestyle="--", color="green",linewidth=1.5)
ax.vlines(x=X5, ymin=Y1, ymax=Y2,linestyle="--", color="red",linewidth=1.5)
ax.vlines(x=X6, ymin=Y1, ymax=Y2,linestyle="--", color="red",linewidth=1.5)
```

```python
ThickmeanB=np.mean(np.concatenate((Th[Y1:Y2,X1:X2],Th[Y1:Y2,X5:X6]),1))
ThickmeanM=np.mean(Th[Y1:Y2,X3:X4])
ThickdestB=np.std(np.concatenate((Th[Y1:Y2,X1:X2],Th[Y1:Y2,X5:X6]),1))
ThickCNR=(np.abs(ThickmeanM-ThickmeanB))/ThickdestB
print("P.Thickness image/The mean Background is",round(ThickmeanB,3))
print("P.Thickness image/The mean Material is",round(ThickmeanM,3))
print("P.Thickness image/The desviation Background is",round(ThickdestB,3))
print("P.Thickness image/The CNR is", round(ThickCNR,3))
```

```python
fig,ax=plt.subplots()
plt.plot(PI(Th,0,Npixeles,0,Npixeles), '-', color="blue")
plt.ylabel("Normalized intensity",fontsize=14)
plt.xlabel("Position (mm)",fontsize=14)
plt.grid()
plt.xlim(0.0,255)
```
